```python
LIMIT = 10000
```

```python
nitrobase_dict = {
    'A' : 0,
    'C' : 1,
    'G' : 2,
    'T' : 3,
}

def input_encoding(x_str):
    values = [nitrobase_dict[b] for b in x_str]
    sz, max_value = len(values), len(nitrobase_dict)
    matrix = np.zeros([sz, max_value], dtype='float32')
    matrix[np.arange(sz), values] = 1
    return matrix
```

```python
import torch
```

```python
for _, row in tqdm.tqdm(some_bp_df_pure.iterrows()):
    bp_id = row['bp_id']
    seq = row['seq']
    x = input_encoding(seq)
    x = torch.tensor(x)
    part = bp_id // LIMIT
    torch.save(x, 'seq_torch_float_many_folders/part_{}/bp_id_{}.pt'.format(part, bp_id))
```
