```python
icgc_files = sorted(list(os.listdir('final_consensus_sv_bedpe/icgc/')))
tcga_files = sorted(list(os.listdir('final_consensus_sv_bedpe/tcga/')))
```

```python
icgc_df_lengths = list()

for fn in tqdm.tqdm(icgc_files):
    df = pd.read_csv(
        os.path.join('final_consensus_sv_bedpe/icgc/', fn), sep='\t'
    )
    icgc_df_lengths.append(len(df))

icgc_df_lengths = np.array(icgc_df_lengths)
```

```python
np.max(icgc_df_lengths), np.mean(icgc_df_lengths), np.median(icgc_df_lengths), np.min(icgc_df_lengths)
```

```python
for i in range(10):
    print(i, ':', np.sum(icgc_df_lengths == i))
```

```python
plt.hist(icgc_df_lengths, bins=50);
```
