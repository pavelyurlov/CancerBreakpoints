```python
f = open('random_sequences.fa.out', 'r')
lines = f.readlines()
```

```python
ID_OFFSET = 1000000
```

```python
bp_id_col = list()
chr_col = list()
start_col = list()
end_col = list()
strand_col = list()
chromo_col = list()
seq_col = list()
```

```python
for i in tqdm.tnrange(len(lines) // 2):
    l = lines[2 * i]
    l = l.strip('>\n)')
    l, strand = l.split('(')
    chr_no, l = l.split(':')
    start, end = l.split('-')

    bp_id_col.append(ID_OFFSET + i)
    chr_col.append(chr_no)
    start_col.append(int(start))
    end_col.append(int(end))
    strand_col.append(strand)
    chromo_col.append(0)
    
    l = lines[2 * i + 1]
    seq_col.append(l.strip())
```

```python
f.close()
```

```python
random_seq_df = pd.DataFrame(data={
    'bp_id' : bp_id_col,
    'chr' : chr_col,
    'start' : start_col,
    'end' : end_col,
    'strand' : strand_col,
    'chromo' : chromo_col,
    'seq' : seq_col
}
                            )
random_seq_df = random_seq_df[
    ['bp_id', 'chr', 'start', 'end', 'strand', 'chromo', 'seq']
]
```
